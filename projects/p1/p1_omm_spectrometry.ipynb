## Exploration


### First, using an Astropy tutorial to view and manipulate the image

```python
file_path = './fit_files/source.fits'
source_file = fits.open(file_path)
source_file.info()

# print(type(source_file))
print(source_file[0])
```

```python
source_data = source_file[0].data
print(type(source_data))
print(source_data.shape)
source_file.close()
```

### Viewing image data and getting basic statistics

```python
# plt.style.use('seaborn-v0_8-white')

plt.imshow(source_data, cmap = 'viridis')
plt.colorbar()

# header = source_file[0].header
# print(repr(header))
```

Let's now get some basic stats about our image

```python
print('Min', np.min(source_data))
print('Max', np.max(source_data))
print('Mean', np.mean(source_data))
print('Stdev', np.std(source_data))
```
